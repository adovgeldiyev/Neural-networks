First shuffle the instances in the original dataset, and take the first 70% as the training and the rest as the test.

```python
import random

random.shuffle(iris_data)
cut_point = int(len(iris_data) * 0.7)
train_set = iris_data[:cut_point]
test_set = iris_data[cut_point:]
```

```python
# Create a new network
net4 = network.Network([4,20,3])
```

Then train the network for 50 epochs with eta = 0.1 and the minibatch size = 5 (and take the default for stopaccuracy). Save the results.

```python
train_net4 = net4.SGD(train_set,50,5, 0.1,test_data=test_set)
```

```python
mse_train=[]
ce_train=[]
loglike_train=[]
for num in train_net4[0]:
    mse_train.append(num[2])
    ce_train.append(num[3])
    loglike_train.append(num[4])
```

```python
mse_test=[]
ce_test=[]
loglike_test=[]
for num in train_net4[1]:
    mse_test.append(num[2])
    ce_test.append(num[3])
    loglike_test.append(num[4])
```

One plot that compares the error curves of the three cost functions for the training set.

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

plt.figure(figsize=(8,6))
plt.plot(mse_train,color='r',label='MSE')
plt.plot(ce_train,color='g',label='CrossEntropy')
plt.plot(loglike_train,color='b',label='LogLikelihood')
plt.title('Error by Cost Function (training)')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend()
plt.show()
```

Three plots, one for each cost function, that shows the error for training vs. test (so that we can inspect for overfitting).

```python
plt.figure(figsize=(8,6))
plt.plot(mse_train,color='r',label='Train')
plt.plot(mse_test,color='y',label='Test')
plt.title('Train vs Test: MSE')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend(prop={'size': 12})
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ce_train,color='r',label='Train')
plt.plot(ce_test,color='y',label='Test')
plt.title('Train vs Test: CrossEntropy')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend(prop={'size': 12})
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(loglike_train,color='r',label='Train')
plt.plot(loglike_test,color='y',label='Test')
plt.title('Train vs Test: LogLikelihood')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend(prop={'size': 12})
plt.show()
```
